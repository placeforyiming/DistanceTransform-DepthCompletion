```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

xx = np.linspace(0, 2 * np.pi, 400)
yy = np.sin(x ** 2)-0.5
plt.close('all')



# row and column sharing
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
ax1.plot(x, y)
ax1.plot(xx, yy)
ax1.set_title('Sharing x per column, y per row')
ax2.scatter(x, y)
ax2.scatter(xx, yy)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')
ax3.scatter(xx, 2 * yy ** 2 - 1, color='r')
ax4.plot(x, 2 * y ** 2 - 1, color='r')
ax4.plot(xx, 2 * yy ** 2 - 1, color='r')


plt.show()
```

```python
error_list_1=[]
f = open("./checkpoints/SparsityCNN/DT/64/False/record.txt", "r")
total_loss=0.0
for i in f:
    each_line=f.readline()
    if not(each_line[0]=='L'):
        total_loss+=float(each_line[10:16])
    else:
        error_list_1.append(total_loss/21.0)
        total_loss=0.0
error_list_1.append(total_loss/21.0)
    
f.close()

error_list_2=[]
f = open("./checkpoints/SparsityCNN/baseline/64/record.txt", "r")
total_loss=0.0
for i in f:
    each_line=f.readline()
    if not(each_line[0]=='L'):
        total_loss+=float(each_line[10:16])
    else:
        error_list_2.append(total_loss/21.0)
        total_loss=0.0
error_list_2.append(total_loss/21.0)
    
f.close()

x=[1,2,3,4,5,6]


error_list_1_x=[]
for i in range(len(error_list_1)):
    if i>0 and i%10==0:
        error_list_1_x.append(float(i)*6.0/len(error_list_1))
    
error_list_1_y_10=[]

for i in range(len(error_list_1)):
    if i>0 and i%10==0:
        error_list_1_y_10.append(np.sum(error_list_1[(i-10):i])/10.0)
plt.scatter(x,np.log(error_list_1[1:]))
plt.plot(x,np.log(error_list_1[1:]))
plt.scatter(x,np.log(error_list_2[1:]))
plt.plot(x,np.log(error_list_2[1:]))
```
